```python
# BIBLIOTECAS
import pandas as pd
import openpyxl
import plotly.express as px
import nbformat
```

```python
#Contar quantas vezes aparece cada correspondência
dados["forma_pagamento"].value_counts()
```

```python
#Agrupando dados - Somar o valor de todas as vendas por loja
dadosagrupados = dados.groupby(["loja","cidade","estado","forma_pagamento"])["preco"].sum().to_frame()
```

```python
#exportar pra excel os dados
dadosagrupados.to_excel("Faturamento.xlsx")
```

```python
#Criar gráficos abançados estilo BI
grafico = px.histogram(dados, x="loja",y="preco",text_auto=True,title="Faturamento",color="forma_pagamento")
grafico.show()
grafico.write_html("faturamento.html")
```
